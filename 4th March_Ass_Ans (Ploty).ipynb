```python
# Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
# scatter plot for age and fare columns in the titanic dataset.
```

```python
import seaborn as sns
import plotly.express as px

# Load the titanic dataset
titanic_df = sns.load_dataset("titanic")

# Create a scatter plot using Plotly Express
fig = px.scatter(titanic_df, x="age", y="fare")


# Show the plot
fig.show()
```

This code will load the titanic dataset using the sns.load_dataset() function from Seaborn and create a scatter plot of the age and fare columns in the titanic_df DataFrame.  The px.scatter() function from Plotly Express is used to create the plot, and the resulting plot will be displayed using the fig.show() method.

You can customize the plot further by adjusting the parameters passed to the px.scatter() function, such as the size or marker type of the points. Additionally, you can save the plot to a file using the fig.write_image() method.

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
```

```python
import plotly.express as px

# Load the tips dataset
tips_df = px.data.tips()


# Create a box plot using Plotly Express
fig = px.box(tips_df, x="day", y="total_bill")

# Show the plot
fig.show()
```

This code will create a box plot of the total_bill column in the tips_df DataFrame, grouped by the day column. The px.box() function from Plotly Express is used to create the plot, and the resulting plot will be displayed using the fig.show() method.

You can customize the plot further by adjusting the parameters passed to the px.box() function, such as the color of the boxes or the orientation of the plot. Additionally, you can save the plot to a file using the fig.write_image() method.

```python
# Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
# the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
# column with the color parameter.
```

```python
import plotly.express as px

# Load the tips dataset
tips_df = px.data.tips()

# Create a histogram using Plotly Express
fig = px.histogram(tips_df, x="sex", y="total_bill", color="day", pattern_shape="smoker")

# Show the plot
fig.show()
```

This code will create a histogram of the total_bill column in the tips_df DataFrame, with separate bars for each value of the sex column. The bars will be grouped by the day column, and the smoker column will be used to display different pattern shapes within each bar. The px.histogram() function from Plotly Express is used to create the plot, and the resulting plot will be displayed using the fig.show() method.

You can customize the plot further by adjusting the parameters passed to the px.histogram() function, such as the number of bins used in the histogram or the opacity of the bars. Additionally, you can save the plot to a file using the fig.write_image() method.

```python
# Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
# the color parameter.
# Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
# dimensions parameter.
```

```python
import plotly.express as px

# Load the iris dataset
iris_df = px.data.iris()

# Create a scatter matrix plot using Plotly Express
fig = px.scatter_matrix(iris_df, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")

# Show the plot
fig.show()
```

This code will create a scatter matrix plot of the sepal_length, sepal_width, petal_length, and petal_width columns in the iris_df DataFrame, colored by the species column. The px.scatter_matrix() function from Plotly Express is used to create the plot, and the resulting plot will be displayed using the fig.show() method.

You can customize the plot further by adjusting the parameters passed to the px.scatter_matrix() function, such as the size or marker type of the points. Additionally, you can save the plot to a file using the fig.write_image() method.

```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.
```

Distplot is a type of plot used to visualize the distribution of a dataset, showing the density of the data over a continuous interval. It is commonly used to display the shape of a distribution, as well as its central tendency, dispersion, and skewness.

Here's an example code to create a distplot using Plotly Express:

```python
import plotly.express as px

# Load the tips dataset
tips_df = px.data.tips()

# Create a distplot using Plotly Express
fig = px.histogram(tips_df, x="total_bill",nbins = 30 , marginal="rug", color_discrete_sequence=["blue"] )


# Show the plot
fig.show()
```

This code will create a distplot of the total_bill column in the tips_df DataFrame, using a histogram to display the density of the data. The px.histogram() function from Plotly Express is used to create the plot, and the resulting plot will be displayed using the fig.show() method.

The nbins parameter controls the number of bins used in the histogram, while the marginal parameter adds a rug plot to the margins of the histogram, showing the individual data points. The color_discrete_sequence parameter sets the color of the histogram bars to blue.

You can customize the plot further by adjusting the parameters passed to the px.histogram() function, such as the bin width or the opacity of the bars. Additionally, you can save the plot to a file using the fig.write_image() method.
